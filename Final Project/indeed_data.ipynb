```python
# scrape job details
all_job_details=[]
for job_soup in tqdm(all_job_soup[10:13], desc="Fetching job details", unit="job"):
    job_title = job_soup.find('h1',{'data-testid':'jobsearch-JobInfoHeader-title'})
    if job_title:
        job_title = job_title.find_next('span').get_text()

    company_tag = job_soup.find('div',{'data-testid':'jobsearch-CompanyInfoContainer'})
    if company_tag:
        company_name = company_tag.find_next('a').get_text()
        company_link = company_tag.find_next('a').get('href')

        job_education = job_soup.find('div',{'role':'group','aria-label':'Education'}).find_next('b').get_text()

        job_details = job_soup.find('div',{'id':'jobDetailsSection'}).find_all('div',{'class':'js-match-insights-provider-16m282m e37uo190'})
        if job_details:
            job_details = [[detail['aria-label'] for detail in job_details],[detail.find('div',{'class':'js-match-insights-provider-tvvxwd ecydgvn1'}).get_text() for detail in job_details]]

        job_description = job_soup.find('div',{'id':'jobDescriptionText'}).get_text()

        job_dict={
            'job_title': [job_title],
            'company_name': [company_name],
            'company_link': [company_link],
            'education': [job_education],
            'pay': [next((job_details[1][i] for i, v in enumerate(job_details[0]) if v == 'Pay'), None)],
            'job_type': [next((job_details[1][i] for i, v in enumerate(job_details[0]) if v == 'Job type'), None)],
            'job_description': [job_description]
        }

        job_df = pd.DataFrame(job_dict)
    
        all_job_details.append(job_df)

print("All job details got!")
```
